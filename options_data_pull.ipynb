Convert date from string to datetime and do datetime calculation to find the option expiration date that is closest to 60 days.  

Also calculate the days to expiration.

```python
type(option_dates_list[0])
```

```python
import datetime

date1 = datetime.datetime.strptime(option_dates_list[0], '%Y-%m-%d')
date1
```

```python
today_date = datetime.datetime.now()
today_date
```

```python
import datetime

target_date = today_date + datetime.timedelta(days=60)
target_date
```

```python
(target_date - today_date).days
```

```python
closest_date = today_date

for x in option_dates_list:
    x = datetime.datetime.strptime(x, '%Y-%m-%d')
    print(x,',',(x-target_date).days)
    if (abs((x - target_date).days)) < abs(((closest_date - target_date).days)):
        closest_date = x

print(closest_date)
days_to_expiration = (closest_date - today_date).days
print(days_to_expiration)
closest_date = closest_date.strftime("%Y-%m-%d")
print(closest_date)
print(type(closest_date))
```

```python
# Use options dates to get the option chain data
# get option chain for specific expiration
# opt = msft.option_chain(option_dates_list[0])
opt = msft.option_chain(closest_date)
# data available via: opt.calls, opt.puts
```
